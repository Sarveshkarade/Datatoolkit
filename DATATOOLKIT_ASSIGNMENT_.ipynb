```python
##Q10:Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import numpy as np
import pandas as pd
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var, name = "Numerical Variable")
sns.histplot(data = num_var, kde = True)
```
